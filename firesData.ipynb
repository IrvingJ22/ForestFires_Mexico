```python
def plot_top_values(data, column, n=10):
   
    frecuencias = data[column].value_counts().head(n)
    
    # Graficar resultados
    plt.figure(figsize=(5, 3))
    frecuencias.plot(kind='bar')
    plt.title(f'Los {n} valores más comunes en la columna {column}')
    plt.xlabel('Valores')
    plt.ylabel('Frecuencia')
    plt.xticks(rotation=90)
    plt.show()
```

# Bayes y probabilidad codicional

### ¿Cuál es la probabilidad de que un incendio sea causado por "Causa específica X" dado que se trata de un "Tipo de incendio Y"?

1. Calcular las probabilidades marginales: Calcula la probabilidad de que un incendio sea de "Tipo de incendio Y" y la probabilidad de que un incendio sea causado por "Causa específica X" de forma independiente.
2. Calcular la probabilidad conjunta: Calcula la probabilidad conjunta de que un incendio sea de "Tipo de incendio Y" y causado por "Causa específica X".
3. Calcular la probabilidad condicional inicial: Calcula la probabilidad condicional inicial de que un incendio sea causado por "Causa específica X" dado que se trata de un "Tipo de incendio Y", utilizando la fórmula del Teorema de Bayes.

```python
# Paso 1: Filtrar el dataframe por "Tipo Vegetación Y"
df_filtrado = df[df['Tipo Vegetación'] == 'bosque de pino - bp']

# Paso 2: Calcular la probabilidad marginal de "Tipo Vegetación Y"
probabilidad_tipo_vegetacion_Y = df_filtrado.shape[0] / df.shape[0]

# Paso 3: Calcular la probabilidad marginal de "Causa específica X"
probabilidad_causa_especifica_X = df[df['Causa especifica'] == 'vandalismo'].shape[0] / df.shape[0]

# Paso 4: Calcular la probabilidad conjunta de "Tipo Vegetación Y" y "Causa específica X"
probabilidad_conjunta = df[(df['Tipo Vegetación'] == 'bosque de pino - bp') & (df['Causa especifica'] == 'vandalismo')].shape[0] / df.shape[0]

# Paso 5: Calcular la probabilidad condicional de "Causa específica X" dado "Tipo Vegetación Y" utilizando el Teorema de Bayes
probabilidad_condicional = (probabilidad_conjunta * probabilidad_tipo_vegetacion_Y) / probabilidad_causa_especifica_X

# Imprimir el resultado
print("La probabilidad de que un incendio sea causado por 'vandalismo' \ndado que la vegetación afectada es de tipo 'bosque de pino - bp' es:", probabilidad_condicional)

```
